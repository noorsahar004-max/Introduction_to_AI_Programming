Electricity consumption

```python
import csv
consumption = []
```

```python
with open("Assignment 1.csv") as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        consumption.append(float(row[1]))
```

```python
total_use = sum(consumption)
avg_use = total_use / len(consumption)
highest_use = max(consumption)
lowest_use = min(consumption)

use_in_kwh = [value / 1000 for value in consumption]
```

```python
print("Total Consumption:",total_use )
print("Average Consumption:", avg_use)
print("Maximum Consumption:", highest_use)
print("Minimum Consumption:", lowest_use)
print("Consumption in kWh:", use_in_kwh)
```
