Computes hourly load and price data from CSV files and calculates the total bill.

```python
import csv

def hourly_bill(loads, prices):
    bills = []
    for i in range(len(loads)):
        bills.append(loads[i] * prices[i])
    return bills

def total(bills):
    return sum(bills)

hrs = []
load = []
price = []

with open("Assignment 3 load.csv") as f:
    r = csv.reader(f)
    next(r)
    for row in r:
        hrs.append(int(row[0]))
        load.append(float(row[1]))

with open("Assignment 3 price.csv") as f:

    r = csv.reader(f)
    next(r)
    for row in r:
        price.append(float(row[1]))

h_bill = hourly_bill(load, price)
t_bill = total(h_bill)

for i in range(len(h_bill)):
    print("Hour", hrs[i], ":", h_bill[i], "Rs")

print("\nTotal Bill:", t_bill, "Rs")
```
