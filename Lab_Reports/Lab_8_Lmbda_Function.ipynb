Lab Report Tasks:

You’re writing small analysis tools for an electrical lab. You need quick one-line functions to calculate adjusted power readings, filter noisy sensor data, sort components by rating, and combine multiple electrical values. Each task builds a specific mini-tool using Python lambda functions.
Task 1 
 Basic lambda for a simple electrical formula
Write a lambda that takes a voltage value V and returns V² + 2.
Test it with V = 5.
(Represents a toy “adjusted voltage metric.”)
Task 2 
 Lambda with two electrical parameters
Write a lambda that takes two resistances R1 and R2 and returns their equivalent parallel resistance using:
R_"eq" =(R1×R2)/(R1+R2)

Test it with:
	R1 = 100 Ω
	R2 = 200 Ω


Task 1

```python
adjusted_voltage = lambda V: V**2 + 2
print(adjusted_voltage(5))
```

Task 2

```python
parallel_resistance = lambda R1, R2: (R1 * R2) / (R1 + R2)
print(parallel_resistance(100, 200))
```
