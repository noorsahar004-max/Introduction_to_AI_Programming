Scenario:
You are designing a small student management system. You need to maintain student data for a class.
 Each student has a name, a roll number, and a list of grades. The system must allow grade updates 
 but keep roll numbers fixed.
 Tasks:
1.	Create a list of student names: ["Ali", "Usman", "Hassan", "Kamran"].
2.	Create a tuple of student roll numbers: (101, 102, 103, 104).
3.	For each student, create a list of grades (3 subjects) using nested lists:
Example: [ [85, 90, 78], [88, 92, 80], ... ]
4.	Update Ali’s second grade to 95.
5.	Add a new grade for Hassan in the fourth subject.
6.	Attempt to change Kamran’s roll number and observe what happens. Explain why.
7.	Print all students' names with their updated grades in a readable format.

```python
names = ["Ali", "Usman", "Hassan", "Kamran"]
roll_numbers = (101, 102, 103, 104)

grades = [
    [85, 90, 78],
    [88, 92, 80],
    [75, 85, 70],
    [90, 88, 92]
]

grades[0][1] = 95
grades[2].append(82)

print(roll_numbers[3] == 104)

for i in range(len(names)):
    print(names[i], ":", grades[i])
```

Kamran’s roll number cannot be changed because roll numbers are stored in a tuple, and tuples are immutable (they do not allow modification).
