Lab 3: Dictionary Comprehension 

Lab Report Tasks:

You are helping a small learning center clean and organize small classroom datasets.
They give you short dictionaries and want quick transformations using comprehensions only, because they want the code short and readable.

Create dictionary from a list
Given:
names = ["Ali", "Sara", "Bilal"]
Create a dictionary where each name maps to its length.
{key: value for item in iterable}

```python
names = ["Ali", "Sara", "Bilal"]
name_lengths = {name: len(name) for name in names}
print(name_lengths)
```

Filter a dictionary using comprehension
Given:
scores = {"Ali": 55, "Sara": 82, "Bilal": 40, "John": 90}
Create a new dictionary containing only the students who scored 60 or above.
{k: v for k, v in dict.items() if condition}

```python
scores = {"Ali": 55, "Sara": 82, "Bilal": 40, "John": 90}
passed = {k: v for k, v in scores.items() if v >= 60}
print(passed)
```

Swap keys and values
Given:
pairs = {"a": 1, "b": 2, "c": 3}
Use a dictionary comprehension to swap keys and values.

```python
pairs = {"a": 1, "b": 2, "c": 3}
swapped = {v: k for k, v in pairs.items()}
print(swapped)
```

Transform dictionary values
Given:
prices = {"pen": 10, "book": 50, "bag": 100}
Use a dictionary comprehension to apply a 10% discount to each price.

```python
prices = {"pen": 10, "book": 50, "bag": 100}
discounted = {k: v * 0.9 for k, v in prices.items()}
print(discounted)
```
