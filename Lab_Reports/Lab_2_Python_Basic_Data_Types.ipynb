Find Total Impedance 
i. Add the three impedances. 
ii. Print real part (resistive) and imaginary part (reactive). 
iii. Print magnitude and phase angle (use abs() and cmath.phase()). 

```python
import cmath

z1 = 5 + 3j
z2 = 2 - 4j
z3 = 3 + 6j

total = z1 + z2 + z3

print(total.real)
print(total.imag)
print(abs(total))
print(cmath.phase(total))
```
